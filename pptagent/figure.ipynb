```python
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ['TextEditing', 'VisualFormatting', 'LayoutAdjustment', 'SlideStructure', 'Overall']
baseline_values = [0.0007, 0.0019, 0.0008, 0.0008, 0.0010]
ui_agent_values = [0.0166, 0.0152, 0.0177, 0.0120, 0.0159]
ours_values = [0.0016, 0.0028, 0.0020, 0.0012, 0.0020]

# Base pastel colors
baseline_color = (200/255, 230/255, 201/255)
ui_agent_color = (255/255, 224/255, 178/255)
ours_color     = (233/255, 243/255, 250/255)

# Vivid versions for 'Overall'
baseline_vivid = (150/255, 200/255, 150/255)
ui_agent_vivid = (255/255, 190/255, 120/255)
ours_vivid     = (180/255, 220/255, 250/255)

# Set font
plt.rc('font', family='Times New Roman')

x = np.arange(len(categories))
width = 0.25
max_val = max(max(baseline_values), max(ui_agent_values), max(ours_values))

fig, ax = plt.subplots(figsize=(10, 4.5))
fontsize_ = 20

# Plot bars with vivid color for 'Overall'
for i in range(len(categories)):
    bc = baseline_vivid if i == 4 else baseline_color
    uc = ui_agent_vivid if i == 4 else ui_agent_color
    oc = ours_vivid if i == 4 else ours_color
    ax.bar(x[i] - width, baseline_values[i], width, color=bc)
    ax.bar(x[i], ui_agent_values[i], width, color=uc)
    ax.bar(x[i] + width, ours_values[i], width, color=oc)

# Format
ax.set_xticks(x)
ax.set_xticklabels(categories, rotation=45, ha='right', fontsize=fontsize_)
ax.set_ylim(0, max_val * 1.1)
ax.set_ylabel('Avg. Cost ($)', fontsize=fontsize_)
ax.legend(['Baseline', 'UI Agent', 'Ours'], fontsize=fontsize_, framealpha=0.5, ncol=3)
ax.tick_params(axis='y', labelsize=fontsize_)

plt.tight_layout()
plt.savefig('cost_fig.pdf', dpi=600)
# plt.show()
```
